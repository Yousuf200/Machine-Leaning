```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
conf_file = "ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model = "frozen_inference_graph.pb"
```

```python
model = cv2.dnn_DetectionModel(frozen_model,conf_file)
```

```python
classLables=[]

filee="Lables.txt"

with open(filee ,"rt")as fpt:
    classLables= fpt.read().rstrip('\n').rsplit("\n")
```

```python
print(classLables)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
immg = cv2.imread("man_test.jpg")
```

```python
plt.imshow(immg)
```

```python
plt.imshow(cv2.cvtColor(immg, cv2.COLOR_BGR2RGB))
```

```python
Classindex , confedence, bbox = model.detect(immg , confThreshold=0.5)
```

```python
print(Classindex)
```

```python
font_size = 3
font = cv2.FONT_HERSHEY_PLAIN

for ClassInd , conf, boxes in zip(Classindex.flatten(), confedence.flatten(), bbox):
    cv2.rectangle(immg,boxes,(255,0,0),2)
    cv2.putText(img=immg, text=classLables[ClassInd-1], org=(boxes[0]+10, boxes[1]+40), fontFace=cv2.FONT_HERSHEY_TRIPLEX, fontScale=3, color=(0, 255, 0),thickness=3)
```

```python
plt.imshow(immg)
```
